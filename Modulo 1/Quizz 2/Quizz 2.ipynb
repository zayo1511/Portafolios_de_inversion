```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tabla = pd.DataFrame({
    'prob': [0.2, 0.3, 0.5],
    'a': [-0.2, 0.05, 0.4],
    'b': [-0.05, 0.1, 0.15],
    'c': [0.05, 0.03, 0.02]
})
```

```python
tabla["porABC"] = 0.4*tabla.a + 0.4*tabla.b + 0.2*tabla.c
tabla
```

```python
ve_porABC = (tabla.prob * tabla.porABC).sum()
np.round(ve_porABC, 3)
```

```python
rendimiento_dinero = 5000 * 0.2 + 22500 * 0.15 + 22500 * 0.17
rendimiento_dinero
```

```python
rendimiento_porcentaje = rendimiento_dinero / 50000
np.round(rendimiento_porcentaje, 3)
```

```python
ea = (tabla.a * tabla.prob).sum()
eb = (tabla.b * tabla.prob).sum()

var_A = (tabla.prob * (tabla.a - ea)**2).sum()
vol_A = var_A**(1/2)

var_B = (tabla.prob * (tabla.b - eb)**2).sum()
vol_B = var_B**(1/2)

cov_AB = (tabla.prob * (tabla.a - ea) * (tabla.b - eb)).sum()

corr_AB = cov_AB / (vol_A * vol_B)
np.round(corr_AB, 3)
```

```python
tabla["porAC"] = 0.3 * tabla.a + 0.7 * tabla.c
ve_porAC = (tabla.prob * tabla.porAC).sum()

var_porAC = (tabla.prob * (tabla.porAC - ve_porAC)**2).sum()
vol_porAC = var_porAC **(1/2)
np.round(vol_porAC, 3), np.round(ve_porAC, 3)
```
