```python
tabla = pd.DataFrame(columns=['prob', 'ret'],
                     index = ["recesión", "normalidad", "expansión"])

tabla['prob'] = [0.4, 0.3, 0.3]
tabla['ret'] = [-0.1, 0.08, 0.23]
tabla
```

```python
valor_esperado = sum(tabla["prob"] * tabla["ret"])
valor_esperado
```

```python
varianza = np.sum(tabla.prob * (valor_esperado - tabla.ret)**2)
volatilidad = varianza**(1/2)
volatilidad
```

```python
import yfinance as yf
import numpy as np

tickers = ['^GSPC']
start_date = '1900-01-01'
end_date = '2023-12-31'
df = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
df = df.resample('Y').last()
rt = df.pct_change()
```

```python
rt = rt.dropna()
rt
```

```python
def rendimiento_aritmetico(x):
    return (1+x).prod() - 1

rendimiento_aritmetico(rt)
```

```python
rg = (1 + rt).prod() ** (1/len(rt)) - 1
rg
```
