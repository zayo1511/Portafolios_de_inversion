## 1

Obtener todos los registros en los cuales **la propina es igual a 10**.

```python
# Solución aquí

tips_data.query("tip == 10")
```

## 2

Obtener todos los registros donde **el número de personas (size) es igual a 5 o 6**.

```python
# Solución aquí

tips_data.query("size == 4 or size == 5")
```

## 3

Obtener todos los registros donde **el grupo contenga fumadores**.

```python
# Solución aquí

tips_data.query("smoker == 'Yes'")
```

## 4

Obtener todos los registros donde **el grupo no contenga fumadores**.

```python
# Solución aquí

tips_data.query("smoker == 'No'")
```

## 5

Obtener todos los registros donde **la propina sea igual a 5, 7 o 10**

```python
# Solución aquí

tips_data.query("tip == 5 or tip == 7 or tip == 10")
```

## 6

Obtenga todos los registros donde **el día sea Sábado o Domingo**.

```python
# Solución aquí

tips_data.query("day == 'Sat' or day == 'Sun'")
```

## 7

Obtenga todos los registros donde **el número de personas sea mayor a 4** y **quien pague la cuenta sea una mujer**.

```python
# Solución aquí

tips_data.query("size > 4 and sex == 'Female'")
```

## 8

Obtenga todos los registros donde **la cuenta total sea mayor a 40** y **la propina sea mayor que 5**

```python
# Solución aquí

tips_data.query("total_bill > 40 and tip > 5")
```

## 9

Obtener todos los registros donde el **número de personas sea mayor que 4** o **el momento de asistencia sea una cena**

```python
# Solución aquí

tips_data.query("size > 4 or time == 'Dinner'")
```

## 10

Obtener todos los registros donde **no hayan fumadores**, la **cuenta total sea mayor a 30**, y **el día sea jueves o viernes**.

```python
# Solución aquí

tips_data.query("smoker == 'No' and total_bill > 30 and (day == 'Thur' or day == 'Fri')")
```

## 11

Obtener todos los registros donde **el día sea viernes o sábado**, el **momento de asistencia sea para la comida**, y el **número de personas sea menor a 4**.

```python
# Solución aquí

tips_data.query("(day == 'Fri' or day == 'Sat') and time == 'Lunch' and size < 4")
```

## 12

### 14 ¿La cuenta total está relacionada con el tamaño? Justifica tu respuesta

```python
tips_data.groupby("size").agg({"total_bill": "mean"})
```

Con esta tabla se puede apreciar la media del total de la cuenta separado por la cantidad de personas en la mesa. Se puede ver que la media de cuenta es proporcional al numero de personas en la mesa, es decir, que mientras más personas hayan en la mesa, más cara será la cuenta.
