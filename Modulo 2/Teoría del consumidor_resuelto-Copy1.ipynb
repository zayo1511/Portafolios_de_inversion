# Imaginen un mundo en donde todo lo que puedan gastar es pizzas o galletas


$$Presupuesto = Precio_{pizza} * Cantidad_{pizza} + Precio_{galletas} * Cantidad_{galletas}$$


Imaginen este caso:

Presupuesto=72

Precio galleta = 6

Pizza = 12

El costo de oportunidad de una pizza es dos galletas

Opciones:

**Si no quiero gastar nada en galletas, puedo comprar 6 pizzas**

**Si no quiero gastar nada en pizzas, puedo comprar 12 galletas**

**Gastar en ambas**

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame()
df["galletas"] = [0, 12]
df["pizzas"] = [6, 0]

plt.plot(df.pizzas, df.galletas)
plt.grid()
plt.xlabel("pizzas")
plt.ylabel("galletas")
plt.title("presupuesto $72")
```

# Imaginemos que el presupuesto sube a $84

**Si no quiero gastar nada en galletas, puedo comprar [7] pizzas**

**Si no quiero gastar nada en pizzas, puedo comprar [14] galletas**

**Gastar en ambas**

```python
## Cree un data frame para un presupuesto de 84, compare con el anterior
df1 = pd.DataFrame()
df1["galletas"] = [0, 14]
df1["pizzas"] = [7, 0]

plt.plot(df1.pizzas, df1.galletas)
plt.grid()
plt.xlabel("pizzas")
plt.ylabel("galletas")
plt.title("presupuesto $84")
```

# Suponemos que el consumidor siempre va a gastar todo su dinero, a mayor gasto mayor utilidad

```python
plt.plot(df1.pizzas, df1.galletas, label = "84")
plt.plot(df.pizzas, df.galletas, label = "72")
plt.grid()
plt.legend()
plt.xlabel("pizzas")
plt.ylabel("galletas")
plt.title("comparación presupuestos")
```
